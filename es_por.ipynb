```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('es_por.csv')
print(df)
```

```python
spain_first = df[(df['Team'] == 'Spain') & (df['Half'] == 'First')]
spain_second = df[(df['Team'] == 'Spain') & (df['Half'] == 'Second')]
portugal_first = df[(df['Team'] == 'Portugal') & (df['Half'] == 'First')]
portugal_second = df[(df['Team'] == 'Portugal') & (df['Half'] == 'Second')]

labels = ['TS', 'SOT', 'xG']

spain_first_values = [int(spain_first['TS'].iloc[0]), int(spain_first['SOT'].iloc[0]), float(spain_first['xG'].iloc[0])]
spain_second_values = [int(spain_second['TS'].iloc[0]), int(spain_second['SOT'].iloc[0]), float(spain_second['xG'].iloc[0])]
portugal_first_values = [int(portugal_first['TS'].iloc[0]), int(portugal_first['SOT'].iloc[0]), float(portugal_first['xG'].iloc[0])]
portugal_second_values = [int(portugal_second['TS'].iloc[0]), int(portugal_second['SOT'].iloc[0]), float(portugal_second['xG'].iloc[0])]

x=np.arange(len(labels))
width=0.2

fig,ax=plt.subplots(figsize=(10,6))
rects1=ax.bar(x-1.5*width, spain_first_values, width, label='Spain First Half')
rects2=ax.bar(x-0.5*width, spain_second_values, width, label='Spain Second Half')
rects3=ax.bar(x+0.5*width, portugal_first_values, width, label='Portugal First Half')
rects4=ax.bar(x+1.5*width, portugal_second_values, width, label='Portugal Second Half')

ax.set_ylabel('Count/xG')
ax.set_title('Spain vs Portugal: Shots and xG by half')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

explanation = "TS: Total Shots   |   SOT: Shots on Target   |   xG: Expected Goals"
plt.text(0, -1.5, explanation, ha='left', va='center', fontsize=10, color='gray')

plt.savefig('shots_metrics.png', dpi=300)

plt.tight_layout
plt.show()
```
